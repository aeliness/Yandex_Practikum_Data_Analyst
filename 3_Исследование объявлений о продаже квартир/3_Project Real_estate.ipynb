## Предобработка данных

### Удаление дубликатов

Начнем с удаления дубликатов. Это связано с тем, что далее у нас будут проводиться расчеты на основании столбца `locality_name`, поэтому сначала нужно привести его в порядок. 

Посмотрим для начала на явные дубликаты.

```python
df.duplicated().sum() #проверка явных дубликатов
```

Явных дубликатов выявлено не было. Проверим столбц, в котором точно могут быть дубликаты из-за разного написания - `locality_name`. 

```python
df['locality_name'].unique() #посмотрим на уникальные значения в столбце
```

```python
df['locality_name'].nunique() #посмотрим на количество всех уникальных значений в столбце
```

В глаза сразу бросается разное написание слова "поселок" (где-то с "ё", где-то с "е"). Также приведем все к одному регистру.

```python
df['locality_name'] = df['locality_name'].str.replace('ё', 'е') #проводим замену букв
df['locality_name'] = df['locality_name'].str.lower() #приводим к нижнему регистру
df['locality_name'].nunique() #проверим количество всех уникальных значений в столбце после обработки
```

#### Вывод

Были удалены дубликаты в столбце `locality_name` посредством замены букв в написании, а также приведения строк к одному регистру. Явных дубликатов в таблице выявлено не было.

### Заполнение пропусков

Посмотрим на количество пропущенных данных в таблице.

```python
df.isna().sum() #считаем количество пропусков в датафрейме
```

```python
df.isna().mean().sort_values(ascending=False)*100 
#посмотрим на количество пропусков в процентном соотношении от общего количества по возрастанию
```

Пропущенные значения в колонках `locality_name ` и `floors_total` составляют менее 0.5%, поэтому мы пренебрежем этими данными и сразу удалим их, на итоги исследования они не повлияют.

```python
df = df.dropna(subset=['locality_name', 'floors_total']).reset_index(drop=True)

#удаляем строки с пропусками в столбцах locality_name и floors_total с заменой индексов
```

Количество пропусков в столбце `is_apartment` составляет 88%. Проверим, какие значения используются в данной колонкеи их тип данных.

```python
df['is_apartment'].unique()
```

Судя по всему, значения, которые используются в данной колонке это 1 или 0 (булев тип данных), 0 - соответствует тому, что данный объект недвижимости не является апартаментами, а 1 - наоборот. Логично предположить, что если в данной колонке имеются пропуски, то недвижимость не является апартаментами. Соответственно, заменим все пропущенные значения на 0. А тип данных приведем к нужному чуть позже.

```python
df['is_apartment'] = df['is_apartment'].fillna(0) #заменим все пропущенные значения на 0
```

Количество пропусков в колонке `balcony` составляет порядка 49%. Можно предположить, что если имеется пропуск данных, то  балконы в квартиры отсутствуют.

```python
df['balcony'] = df['balcony'].fillna(0) #заменим все пропущенные значения на 0
```

Далее имееются большое количество пропусков в столбцах:
* `parks_nearest`, 
* `ponds_nearest`, 
* `parks_around3000`,
* `ponds_around3000`,
* `airports_nearest`, 
* `cityCenters_nearest`,

Значения в столбцах должны формировать автоматически на основании картографических данных (кроме столбцов с указанием количества парков и прудов). Возможно, это не произошло из-за какого-то технического сбоя или еще по каким-то причинам.

Пропуски в колонках `parks_around3000` и `ponds_around3000`, вероятнее всего, связаны с тем, что при заполнении ячеек о точном количестве было неизвестно, соответственно, данные поля решили оставить пустыми.

Для этих колонок рассчитаем медианное значение по каждому столбцу в зависимости от их местоположения. 

```python
#напишем цикл, который будет рассчитывать медианные расстояния нужной колонки в зависимости от местоположения, а далее заменять
#пропущенные значения нужной колонки на рассчитанное медианное значение
for index in df['locality_name'].unique(): 
    median = df.loc[df['locality_name'] == index, 'airports_nearest'].median()
    df.loc[(df['airports_nearest'].isna()) & (df['locality_name'] == index), 'airports_nearest'] = median

for index in df['locality_name'].unique():
    median = df.loc[df['locality_name'] == index, 'cityCenters_nearest'].median()
    df.loc[(df['cityCenters_nearest'].isna()) & (df['locality_name'] == index), 'cityCenters_nearest'] = median
  
for index in df['locality_name'].unique():
    median = df.loc[df['locality_name'] == index, 'parks_nearest'].median()
    df.loc[(df['parks_nearest'].isna()) & (df['locality_name'] == index), 'parks_nearest'] = median
    
for index in df['locality_name'].unique():
    median = df.loc[df['locality_name'] == index, 'ponds_nearest'].median()
    df.loc[(df['ponds_nearest'].isna()) & (df['locality_name'] == index), 'ponds_nearest'] = median    

for index in df['locality_name'].unique():
    median = df.loc[df['locality_name'] == index, 'parks_around3000'].median()
    df.loc[(df['parks_around3000'].isna()) & (df['locality_name'] == index), 'parks_around3000'] = median   

for index in df['locality_name'].unique():
    median = df.loc[df['locality_name'] == index, 'ponds_around3000'].median()
    df.loc[(df['ponds_around3000'].isna()) & (df['locality_name'] == index), 'ponds_around3000'] = median   
        
    
    
print('Оставшееся количество пропусков после обработки в колонке airports_nearest:', df['airports_nearest'].isna().sum())  
print('Оставшееся количество пропусков после обработки в колонке cityCenters_nearest:', df['cityCenters_nearest'].isna().sum())
print('Оставшееся количество пропусков после обработки в колонке parks_nearest:', df['parks_nearest'].isna().sum())
print('Оставшееся количество пропусков после обработки в колонке ponds_nearest:', df['ponds_nearest'].isna().sum())
print('Оставшееся количество пропусков после обработки в колонке parks_around3000:', df['parks_around3000'].isna().sum())
print('Оставшееся количество пропусков после обработки в колонке ponds_around3000:', df['ponds_around3000'].isna().sum())
```

Вероятнее всего, пропуски остались из-за того, что невозможно расчитать медиану по определенному населенному пункту, потому что данные по нему отстутствует. 
В данном конкретном случае не остается ничего, кроме как заменить пропуски медианой по столбцу. 

```python
#заменяем пропущенные значения в колонке медианным значением в целом по столбцу
df.loc[df['airports_nearest'].isna(), 'airports_nearest'] = df['airports_nearest'].median() 
df.loc[df['cityCenters_nearest'].isna(), 'cityCenters_nearest'] = df['cityCenters_nearest'].median()
df.loc[df['parks_nearest'].isna(), 'parks_nearest'] = df['parks_nearest'].median()
df.loc[df['ponds_nearest'].isna(), 'ponds_nearest'] = df['ponds_nearest'].median()
df.loc[df['parks_around3000'].isna(), 'parks_around3000'] = df['parks_around3000'].median()
df.loc[df['ponds_around3000'].isna(), 'ponds_around3000'] = df['ponds_around3000'].median()
```

Далее обработаем пропуски в столбце `ceiling_height` (высота потолков), их заменим медианным значением в столбце. 

```python
#заменяем пропущенные значения в колонке медианным значением в целом по столбцу
df.loc[df['ceiling_height'].isna(), 'ceiling_height'] = df['ceiling_height'].median()

df.sample(10) #выведем рандомные 10 строк таблицы для проверки замен
```

Так же были выявлены пропуски в колонках с площадью кухни и жилой площадью. Пропущенные данные заменим медианным значением площадей (кухни и жилой) в зависимости от количества комнат в квартиры.

```python
#для начала проверим, не является ли квартира - квартирой-студией, потому что, если там имеется пропуск в площади кухни,
#то логично, что площадь кухни будет 0, потому что она объединена с комнатой
df.loc[df['studio'] == 1, 'kitchen_area'] = 0 

#а далее напишем цикл, который будет заменять пропуски медианным значением в зависимости от количества комнат в квартире
for index in df['rooms'].unique():
    median = df.loc[df['rooms'] == index, 'kitchen_area'].median()
    df.loc[(df['kitchen_area'].isna()) & (df['rooms'] == index), 'kitchen_area'] = median

#проверим остались ли пропуски    
df['kitchen_area'].isna().sum() 
```

```python
#повтор цикла для жилой площади
for index in df['rooms'].unique():
    median = df.loc[df['rooms'] == index, 'living_area'].median()
    df.loc[(df['living_area'].isna()) & (df['rooms'] == index), 'living_area'] = median
    
df['living_area'].isna().sum()
```

## Расчёты и добавление результатов в таблицу

В соответствии с заданием необходимо рассчитать и добавить в таблицу следующие данные: 
* стоимость 1 кв. м квартиры;
* день недели, месяц и год публикации объявления;
* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
#создадим новый столбец "стоимость за 1 кв. м"
df['price_per_metre'] = round(df['last_price'] / df['total_area'])
df.sample(5) #выведем рандомные 5 строк таблицы
```

```python
#создадим новые столбцы с днем, месяцем и годом размещения объявления
df['day_of_posting'] = df['first_day_exposition'].dt.weekday
df['month_of_posting'] = df['first_day_exposition'].dt.month
df['year_of_posting'] = df['first_day_exposition'].dt.year
df.sample(5) #выведем рандомные 5 строк таблицы
```

```python
#создадим фукцию, которая будет определять этаж расположения квартиры и добавлять ее в соответствующую группу
def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'
        
df['floor_group'] = df.apply(floor_group, axis=1)
df.sample(5)
```

```python
df['living_to_total'] = round(df['living_area'] / df['total_area'],2) #посчитаем соотношение жилой площади к общей с округлением
df['kitchen_to_total'] = round(df['kitchen_area'] / df['total_area'],2) #посчитаем соотношение площади кухни к общей с округлением
df.sample(5)
```

#### Выводы

Были успешно рассчитаны следующие показатели и добавлены новые столбцы: стоимость за 1 кв. м, день недели, месяц и год публикации объявления, сделана группировка этажности (первый, последний, другой), посчитатано сооотношение жилой площади и площади кухни к общей.

### Определение центра Санкт-Петербурга 

```python
#добавим новый столбец cityCenters_km, в котором будут указно расстояние до центра города в километрах, 
#а не метрах (сразу с округлением до целых)
df_good = df_good.assign(cityCenters_km = lambda x: (x['cityCenters_nearest'] / 1000).astype(int)) 
df_good.sample(5)
```

```python
#создадим новую таблицу, в которую включим только объекты, расположенные в Санкт-Петербурге, 
#а также добавим туда только необходимые нам столбцы

spb = (
    df_good
    .query('locality_name in ("санкт-петербург")')
    .loc[:, ('cityCenters_km', 'last_price', 'total_area', 'floor_group', 'rooms', 'ceiling_height', 'day_of_posting', 'month_of_posting', 'year_of_posting')]

)
```

```python
#создадим сводную по расстоянию до центра и стоимости недвижимости
spb_km = spb.pivot_table(index='cityCenters_km', values='last_price')
#построим график, чтобы посмотреть как эти параметры коррелируют между собой
spb_km.plot(style='o-', figsize=(10,5), grid=True)
```

```python
#выберем интересубщее нас расстояние
spb_km.plot(style='o-', figsize=(10,5), xlim=(2,10), grid=True)
```

Судя по графикам, на стоимость квартир, расположенных в пределах 4 км от центра, имеет влияение какие-то еще факторы. После 4 км цена начинает планомерно снижаться вполть до 9 км, после этого происходит некое стагнирование стоимости. 

**Поэтому будет предпологать, что центром считается расстояние до 9 км.**

### Определение ценообразующих факторов для квартир, расположенных в центре Санкт-Петербурга

```python
#создадим новую таблицу, в которую включим только объекты, расположенные в Санкт-Петербурге в пределах 9 км от центра, 
#а также добавим туда только необходимые нам столбцы
spb_centre = (
    spb
    .query('cityCenters_km <= 9')
    .loc[:, ('cityCenters_km', 'last_price', 'total_area', 'rooms', 'floor_group', 'ceiling_height', 'day_of_posting', 'month_of_posting', 'year_of_posting')]

)

spb_centre.sample(5)
```

```python
#посмотрим статистическое описание столбцов
round(spb_centre[['total_area', 'last_price', 'rooms', 'ceiling_height']].describe(), 2) 
```

Медианные и средние значения для квартир, расположенных в пределах 7 км от центра Санкт-Петербурга почти не отличаются. Сверим показатели со всей выборкой. 

```python
#посмотрим статистическое описание столбцов
round(df_good[['total_area', 'last_price', 'rooms', 'ceiling_height']].describe(), 2)
```

Квартиры, продающиеся в центре Санкт-Петербурга отличаются несколько большей площадью, а также более высокой ценой (разница около 40% по медианному значению). Число комнат и высота потолков почти не отличаются.

```python
print('Коэффициент корреляции составляет:', round(spb_centre['last_price'].corr(spb_centre['rooms']), 2))

#построим гистрограмму 
spb_centre['rooms'].plot(kind='hist', bins=10, range=(0,10), grid=True, title='Зависимость цены от количества комнат', figsize=(10,5), legend=True) 
plt.xlabel('кв.м')
plt.ylabel('Количество объявлений')
```

Коэффициент корреляции (0,48) показывается немного более высокую зависимость цены от количества комнат квартир в центре, чем всех остальных (0,41). 

Посмотрим зависимость цены от этажа.

```python
print('Коэффициент корреляции цена-первый этаж составляет:', round(spb_centre['last_price'].corr(spb_centre['floor_group'] == 'первый'),2))
print('Коэффициент корреляции цена-последний этаж составляет:', round(spb_centre['last_price'].corr(spb_centre['floor_group'] == 'последний'),2))
print('Коэффициент корреляции цена-другой этаж составляет:', round(spb_centre['last_price'].corr(spb_centre['floor_group'] == 'другой'),2))

#проверим зависимость стоимости квартиры от количества комнат

#построим гистрограмму 
spb_floor = spb_centre.pivot_table(index='floor_group', values='last_price')
spb_floor.plot(kind='bar', legend=False)
plt.title('Зависимость цены от этажа')
plt.xlabel('Этаж')
plt.ylabel('Цена, млн руб.')
```

Как и в случае с квартирами, расположенными не в центре Санкт-Петербурга, стоимость квартир на первом этаже ниже, чем у остальных. При этом у квартир, расположенных в центре, нахождение на последнем этаже почти не снижает стоимость.

Далее проверим зависимость стоимости от расстояния до центра.

```python
print('Коэффициент корреляции составляет:', round(spb_centre['last_price'].corr(spb_centre['cityCenters_km']), 2)) #проверим зависимость стоимости квартиры от количества комнат


spb_centre_km = spb_centre.pivot_table(index='cityCenters_km', values='last_price')
spb_centre_km.plot(style='o-', figsize=(10,5), grid=True)
```

Корреляция стоимости от удаленности от центра, как и в общем случае, отрицательная. Чем дальше от центра, тем дешевле.

Посмотрим, есть ли зависимость от даты размещения объявления.

```python
#построим столбчатые диаграммы для дня, месяца и года размещения объявления
days_centre = spb_centre.pivot_table(index='day_of_posting', values='last_price', aggfunc='mean')
days_centre.plot(y='last_price', kind='bar', legend=False)
plt.show()

month_centre = spb_centre.pivot_table(index='month_of_posting', values='last_price', aggfunc='mean')
month_centre.plot(y='last_price', kind='bar', legend=False)
plt.show()

year_centre = spb_centre.pivot_table(index='year_of_posting', values='last_price', aggfunc='mean')
year_centre.plot(y='last_price', kind='bar', legend=False)
plt.show()
```

Зависимости стоимости от дня выставления на продажу выявлено не было. При этом, судя по графикам,  стоимость квартир, расположенных в центре, вырастает в сентябре (по общей выборке небольшой рост был в апреле). В обеих выборках стоимость немного снижается в июне, что, видимо, связано с началом сезона отпусков. 

При этом, за период 2016-2019 гг. можно сказать, что стоимость квартир в центре почти не изменялась, в отличие от всех остальных, где в 2019 г. был зафиксирован рост.

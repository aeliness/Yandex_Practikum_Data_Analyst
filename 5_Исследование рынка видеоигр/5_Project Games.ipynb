# Предобработка данных

## Стиль заголовков

Для начала приведем все наименования столбцов к нижнему регистру. 

```python
data.columns = data.columns.str.lower() #приведем названия столбцов к нижнему регистрц
data.sample() #выведем рандомную строку для проверки переименования столбцов
```

## Заполнение пропусков

Далее посмотрим количество пропусков в таблице. 

```python
data.isna().sum() #посчитаем количество пропусков в датафрейме
```

Больше всего пропусков в колонках с оценками и рейтингом. Вероятнее всего, это связано с тем, что данные игры не имееют достаточного количества оценок от пользователей/критиков, поэтому там образовались пропуски, т.е. данные пропуски являются не случайными. 
Что касается рейтинга, то в данных используется рейтинг компании ЕSRB, которая осуществляет свою деятельность на территории США и Канады. Помимо ЕSRB есть еще несколько крупных компаний, занимающихся присвоением рейтинга:
* Pan European Game Information (PEGI) - Европа, Квебек, Россия;
* Unterhaltungssoftware Selbstkontrolle (USK) - Германия,
* Computer Entertainment Rating Organization (CERO) - Япония, 
* Australian Government Classification Board (AGCB) - Австралия,
* Russian Age Rating System (RARS) - Россия. 

В каких-то странах имеются строгие требования к присвоению рейтинга именно той организации, которая осуществляет свою деятельность на территории выпуска игры, в каких-то нет, все зависит от законодательства. Например, в России играм присваивают рейтинги и ESRB, и PEGI, и RARS. Все зависит именно от степени странового охватат игр.
Исходя из этого, можно сделать вывод, что часть игр не имеют рейтинга ESRB, а имеют рейтинг от какой-либо другой организации. Данный тип пропусков не является случайным. 

Также имееются пропущенные данные в наименовании игр, жанре и годе выпуска. Ввиду их небольшого количества, эти данные можно удалить.

```python
#удалим строки с пропусками в столбцах name и genre с заменой индексов
data = data.dropna(subset=['name', 'genre', 'year_of_release']).reset_index(drop=True)
```

```python
data.isna().sum() #проверим количество пропусков
```

Из обзора данных знаем, что в пользовательских оценках присутствует аббревиатура tbd (вероятнее всего означающая, что оценка будет проставлена позже). Посмотрим общее количество таких значений. 

```python
data[data['user_score'] == 'tbd']['user_score'].count()
```

Ввиду того, что данное значение не является случайным, мы заменим его на Nan (в том числе, чтобы можно было в дальнейшем привести данные в другой тип данных). Заменять в дальнешем их ни на что не будем, чтобы эти значения не повлияли на дальнейшие расчеты.  

Пропущенные значения в колонке raiting заменим на unknown. Заменять какими-то другими значениями не будем, поскольку присвоение рейтинга сильно зависит от страны, в которой игра будет выпущена (разные требования к ограничению по возрасту)

```python
data.loc[data['user_score'] == 'tbd','user_score'] = np.nan #заменяем значения tbd на тфт
data[data['user_score'] == 'tbd']['user_score'].count() #подсчитаем количество tbx в таблице
```

```python
#заменяем пустные значения в колонке rating на unknown
data['rating'] = data['rating'].fillna('unknown')
data.isna().sum() #проверим остались ли пропущенные значения
```

Все пропуски были успешно обработаны.

## Изменение типов данных

```python
data.dtypes #посмотрим информацию на типах данных в таблице
```

Год выпуска игры заменим на целочисленный тип данных. Колонки с данными по количеству выпущенных игр, а также с оценками критиков и пользователей заменим на float32 для экономия места.

```python
#замена данные на целочисленный тип int
data['year_of_release'] = data['year_of_release'].astype('int')
#замена данные на вещественный тип float32
data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'critic_score','user_score']] = \
    data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'critic_score','user_score']].astype('float32')

data.info() #выведем общую информацию о датафрейме
```

Тип данных был успешно изменен.

Исходя из данных сводной таблицы, а также гистограммы, можно сделать вывод, что пик продаж приходился 2008 и 2009 гг. -    1 427 млн и 1 426 млн штук соответственно. Далее посмотрим количество проданных игр по платформам.

```python
#строим сводную таблицу с количеством проданных игр по платформам
data_platform = data.pivot_table(index = 'platform', values='total_sales', aggfunc='sum')

#отсортируем таблицу по возрастанию продаж и выведем первые 10
data_platform = data_platform.sort_values(by='total_sales', ascending=False).reset_index()
data_platform
```

Как видно из таблицы выше наиболее популярными являются первые 6 платформ: PS2, X360, PS3, Wii, DS, PS - по ним было продано максимальное количество игр. Далее посмотрим распределение по годам для этих платформ.

```python
top_platform =  list(data_platform['platform'][:6]) #создадим список по платформам (топ-6 по продажам)

#отсортируем таблицу по топ-6 платформам и посмотрим сводную по годам по объему проданных копий и построим диаграмму
data.query('platform == @top_platform')\
.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')\
.plot(kind='bar', grid=True, figsize=(12,5))
```

На диаграмме можно увидеть, что примерный срок службы одной платформы составляет порядка 10 лет. При этом, Sony при выпуске новой PlayStation поддерживает игры на предыдущую модель, как минимум, 4-5 лет. После запуска новой модели продажи на нее начинают опережать продажи на предыдущую версию на 2-3 году жизни.

```python
# отфильтруем таблицу по году выхода игры (с 2015 г.) и построим сводную в разрезе платформ и проданных копий за 2 года
data_actual_sales = data.query('year_of_release >= 2015')\
.pivot_table(index='platform', columns='year_of_release', values='total_sales', aggfunc = 'sum')
data_actual_sales
```

# Проверка гипотез

Далее проверим две гипотезы: 
1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые,
2. Средние пользовательские рейтинги жанров Action и Sports разные.

## Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем гипотезы:

* H0 - средние пользовательские рейтинги платформ Xbox и PC одинаковые;
* H1 - средние пользовательские рейтинги платформ Xbox и PC разные.

Для начала нам необходимо будет отфильтровать датафреймы по платформам, а также убрать оттуда все оценки равные -1 (бывшие пропуски  и tbd в таблице), чтобы они не портили статистику. Далее рассчитаем дисперсию получившихся генеральных совокупностей и посмотрим насколько они отличаются.

```python
#отфильтруем таблицу по платформам и пользовательским оценкам
df_xbox = data_actual_years_2016.query('platform == "XOne" and user_score >=0')
df_pc = data_actual_years_2016.query('platform == "PC" and user_score >=0')

#создадим отдельные переменные, которые будут хранить пользовательские оценки для определенных платформ
df_xbox_score = df_xbox['user_score']
df_pc_score = df_pc['user_score']

#рассчитаем соотношение дисперсий
variance = np.var(df_xbox_score) / np.var(df_pc_score)
display(variance)
```

Дисперсии генеральных совокупностей не равны, соответственно, дальше в расчетах применим параметр equal_var=False. Гипотезу будет проверять с помощью t-теста, т.к. мы делаем предположение о равенстве средних двух независимых совокупностей.

```python
alpha = .05 #зададим критерий статистической значимости 
            #если p-value окажется меньше него - отвергнем гипотезу

#проверим гипотезу о равенстве двух независимых совокупностей
results = st.ttest_ind(df_xbox_score, df_pc_score, equal_var = False)

print('р-значение:', results.pvalue)
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не отвергаем нулевую гипотезу')
```

Значение статистической значимости превышает критический уровень статистической значимости, поэтому мы не отвергаем нулевую гипотезу: средние пользовательские рейтинги платформ XOne и PC могут быть равны.

```python
display(df_xbox_score.mean())
display(df_pc_score.mean())
```

Как видно из данных выше, средние пользовательские рейтинги и правда примерно равны.

## Средние пользовательские рейтинги жанров Action и Sports разные.


Сформулируем гипотезы:

* H0 - cредние пользовательские рейтинги жанров Action и Sports одинаковые;
* H1 - cредние пользовательские рейтинги жанров Action и Sports разные.

```python
#отфильтруем таблицу по жанрам и пользовательским оценкам
df_action = data_actual_years_2016.query('genre == "Action" and user_score >=0')
df_sports = data_actual_years_2016.query('genre == "Sports" and user_score >=0')

#создадим отдельные переменные, которые будут хранить пользовательские оценки для определенных жанров
df_action_score = df_action['user_score']
df_sports_score = df_sports['user_score']

#рассчитаем соотношение дисперсий
variance = np.var(df_action_score) / np.var(df_sports_score)
display(variance)
```

Дисперсии генеральных совокупностей не равны, соответственно, дальше в расчетах применим параметр equal_var=False. Гипотезу будет проверять с помощью t-теста, т.к. мы делаем предположение о равенстве средних двух независимых совокупностей.

```python
alpha = .05 #зададим критерий статистической значимости 
            #если p-value окажется меньше него - отвергнем гипотезу

#проверим гипотезу о равенстве двух независимых совокупностей
results = st.ttest_ind(df_action_score, df_sports_score, equal_var = False)

print('р-значение:', results.pvalue)
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не отвергаем нулевую гипотезу')
```

Значение статистической значимости меньше критического уровеня статистической значимости, поэтому мы отвергаем нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sport не равны.

```python
display(df_action_score.mean())
display(df_sports_score.mean())
```

Как видим, средние пользовательские оценки жанров Action и Sport действительно отличаются на 30%.
